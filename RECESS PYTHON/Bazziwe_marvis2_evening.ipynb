```python
# comparing the elements of two Pandas series. 
import pandas as pd
series1 = pd.Series([4, 65, 436, 3, 9])
series2 = pd.Series([7, 0, 3, 897, 9])
print(series1 == series2)
```

```python
#Pandas program to add, subtract, multiply and divide two Pandas series. 
import pandas as pd
series1 = pd.Series([2, 4, 6, 8, 14])
series2 = pd.Series([1, 3, 5, 7, 9])
print(series1 + series2)
print(series1 - series2)
print(series1 * series2)
print(series1 / series2)
```

```python
# convert a dictionary to a Pandas series
import pandas as pd
dictionary1 = {'Josh': 24, 'Sam': 36, 'Peace': 19, 'Charles': 65, 'Tom': 44}
series1 = pd.Series(dictionary1)
print(series1)
```

```python
#convert a given series to an array
import pandas as pd
series1 = pd.Series(["Love", 800, "Joy", 789.9, "Peace", True])
array1 = series1.to_numpy()
print(array1)
```

```python
# display the most frequent value in the given series and replace everything else as ‘Other’ in the series
import pandas as pd
df_goal = pd.read_csv('AfricaCupofNationsMatches.csv')
home_team_goals = df_goal['HomeTeamGoals']
most_frequent_value = home_team_goals.mode()[0]
home_team_goals.replace(most_frequent_value, 'Other', inplace=True)
print(home_team_goals)
```
